```python
def mann_whitney_func (r1, r2, alpha=0.05):
    w, p_value=stats.mannwhitneyu(r1, r2, alternative = 'two-sided')
    if p_value < alpha:
        print('Распределения не равны')
    else:
        print('Распределения равны')
        
    return w, p_value
```

```python
mann_whitney_func (df[df['id_group']==0]['nflag_order'], df[df['id_group']==1]['nflag_order'], alpha=0.05)
```
